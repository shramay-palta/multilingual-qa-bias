###Sentence Embeddings

```python
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)
```

```python
sentences = ['This is an example sentence', 'Each sentence is converted']

# Load model from HuggingFace Hub
# tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
# model = AutoModel.from_pretrained('sentence-transformers/xlm-r-100langs-bert-base-nli-stsb-mean-tokens')
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/distiluse-base-multilingual-cased-v2')
model = AutoModel.from_pretrained('sentence-transformers/distiluse-base-multilingual-cased-v2')

# Tokenize sentences
encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# Compute token embeddings
with torch.no_grad():
    model_output = model(**encoded_input)

# Perform pooling. In this case, max pooling.
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

print("Sentence embeddings:")
print(sentence_embeddings)
```

```python
questions = defaultdict(list)
for datum in train_data['data']:
  for para in datum['paragraphs']:
    for qa in para['qas']:
      lang = qa['id'].split("-")[0]
      if lang == "english":
        questions["en"].append(qa['question'])
      elif lang == "russian":
        questions["ru"].append(qa['question'])
```

```python
for q in questions:
  print(q, len(questions[q]))
```

```python
sent_embed = defaultdict(list)
for lang, sentences in questions.items():
  sentences = sentences[:500]
  encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

  # Compute token embeddings
  with torch.no_grad():
      model_output = model(**encoded_input)

  # Perform pooling. In this case, max pooling.
  sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
  sent_embed[lang] = sentence_embeddings
```

```python
X = np.array([emb.tolist() for emb in sent_embed['en']] + [emb.tolist() for emb in sent_embed['ru']] )
X_embedded = TSNE(n_components=2, learning_rate='auto', init='random').fit_transform(X)
y = np.array([0 for i in range(len(sent_embed['en']))] + [1 for i in range(len(sent_embed['en']))])
```

```python
figure(figsize=(8, 6), dpi=80)
scatter = plt.scatter(X_embedded[:,0], X_embedded[:,1], c=y)
handles, _ = scatter.legend_elements(prop='colors')
plt.legend(handles, ["EN", "RU"], fontsize=16)
plt.title("tSNE projection of question embeddings", fontsize=20)
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/CMSC848Q/tydiqa/fig/en_ru_tsne_distiluse.pdf')
```

```python
candidate_pairs = []
cos_result = np.zeros((len(sent_embed['en']), len(sent_embed['ru'])))
for en_idx, en_emb in enumerate(sent_embed['en']):
  for ru_idx, ru_emb in enumerate(sent_embed['ru']):
    cos_sim = 1 - spatial.distance.cosine(en_emb, ru_emb)
    cos_result[en_idx, ru_idx] = cos_sim
    candidate_pairs.append((en_idx, ru_idx, cos_sim))
```

###Translation

```python
ru2en_tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-ru-en")
ru2en_model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-ru-en")
# ru2en_model = ru2en_model.to('cuda')
```

```python
print(ru2en_tokenizer)
```

```python
all_src_text = questions['ru'][:500]
translated = []
i = 0
while i <= 450:
  src_text = all_src_text[i:i+50]
  encoded_input = ru2en_tokenizer(src_text, padding=True, truncation=True, return_tensors='pt') #.to('cuda')
  print(encoded_input)
  # with ru2en_tokenizer.as_target_tokenizer():
  #   labels = ru2en_tokenizer(tgt_texts, ...)
  with torch.no_grad():
      model_output = ru2en_model.generate(**encoded_input)
  # translated = [tokenizer.decode(t, skip_special_tokens=True) for t in model_output]
  out = ru2en_tokenizer.batch_decode(model_output, skip_special_tokens=True)
  translated.extend(out)
  i += 50

# model_inputs = tokenizer(src_texts, ...)
# with tokenizer.as_target_tokenizer():
#     labels = tokenizer(tgt_texts, ...)
# model_inputs["labels"] = labels["input_ids"]
```

```python
from nltk.tokenize import word_tokenize
from nltk.translate.bleu_score import sentence_bleu
reference = [['this', 'is', 'a', 'test'], ['this', 'is' 'test']]
candidate = ['this', 'is', 'a', 'test']
score = sentence_bleu(reference, candidate)
print(score)
```

```python
import csv
from nltk.translate.bleu_score import SmoothingFunction  

header = ['en_question', 'ru_question', 'ru_to_en', 'cos_sim', 'bleu']
# data = [[questions['en'][e], questions['ru'][r],c] for e, r, c in candidate_pairs]
data = []
for e, r, cos_sim in candidate_pairs:
  en_question = questions['en'][e]
  ru_question = questions['ru'][r]

  ref = word_tokenize(en_question)
  tra = word_tokenize(translated[r])
  bleu = sentence_bleu(ref, tra, smoothing_function=SmoothingFunction().method1)
  data.append([en_question, ru_question, translated[r], cos_sim, bleu])
  

with open('/content/drive/MyDrive/CMSC848Q/tydiqa/en_ru_sim_pairs_distiluse_0418.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    # write the data
    writer.writerows(data)
```
